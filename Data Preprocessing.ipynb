```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('gsoc-data-cleaned.csv')
data.set_index('Unnamed: 0', inplace=True)
data.head(10)
```

```python
data.info()
```

```python
data.drop(columns=['logo_bg', 'logo_image', 'technologies', 'topics'], axis=1, inplace=True)
data.info()
```

```python
# Cleaning up the string values
for colname in data.columns:
    print(colname)
    if colname=='year' or colname=='num_projects':
        continue
    data[colname] = data[colname].apply(lambda x: str(x).strip())
```

```python
data.to_csv('gsoc-data-cleaned.csv')
```
